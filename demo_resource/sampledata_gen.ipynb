```python
# -*- coding: utf-8 -*-

def genSampleTask():
    import datetime
    from datetime import timedelta
    def setDaytimeRange(slot_daytime, st_daytime, en_daytime):
        '''
        Enter time as argument. 24 hour notation and 0.5 for xx:30 time.
        It will set those given range to 1. en_daytime to en_daytime+30min timeslot is not included.
        
        '''
        new_slot_daytime = [0] * 48
        
        if(st_daytime - int(st_daytime) > 0):
            # Handle xx:30
            index_st = int(st_daytime) * 2 + 1
        else:
            index_st = int(st_daytime) * 2
            
        if(en_daytime - int(en_daytime) > 0):
            # Handle xx:30
            index_en = int(en_daytime) * 2 + 1
        else:
            index_en = int(en_daytime) * 2
            
        new_slot_daytime = slot_daytime
        new_slot_daytime[index_st:index_en] = [1] * (index_en - index_st)
        
        return new_slot_daytime
    
    def createTaskInfo(user_id, task_name, duedate):
        '''
        user_id is ObjectID. not str. 
        
        '''
        obj = {};
        obj["userId"] = user_id
        obj["name"] = task_name
        obj["importance"] = 0
        obj["state"] = 100
        obj["timePreferenceScore"] = [0] * (48*7)
        
        if(duedate != None):
            obj["duedate"] = duedate
        else:
            obj["duedate"] = ''
            
        obj['processedtime'] = 0
        
        return obj;
              
    
    task_collection = [];
    
    # reference should be Sunday, midnight 
    date_reference = datetime.datetime(2015, 11, 15, 0, 0)
    
    # Generate tasks for one week from the date_reference.
    # Sun
    task_name = u'멘토님과 점심 식사'
    duedate = date_reference + datetime.timedelta(days=0, hours=13)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'웹프로그래밍 스터디 자료 준비'
    duedate = date_reference + datetime.timedelta(days=0, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'프로그래밍언어구조론 과제'
    duedate = date_reference + datetime.timedelta(days=0, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Mon
    task_name = u'일주일 계획 세우기'
    duedate = date_reference + datetime.timedelta(days=1, hours=1)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'자전거 타기'
    duedate = date_reference + datetime.timedelta(days=1, hours=11)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'프로그래밍언어구조론 프로젝트 제안서 제출 마감'
    duedate = date_reference + datetime.timedelta(days=1, hours=16)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Tue
    task_name = u'논리회로설계 시험 공부'
    duedate = date_reference + datetime.timedelta(days=2, hours=15)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'작업 추천 알고리즘 구상'
    duedate = date_reference + datetime.timedelta(days=2, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'멘토링 자료 준비'
    duedate = date_reference + datetime.timedelta(days=2, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Wed
    task_name = u'교수님 면담'
    duedate = date_reference + datetime.timedelta(days=3, hours=13)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'멘토링 자료 준비'
    duedate = date_reference + datetime.timedelta(days=3, hours=18)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'음반 예약 구매'
    duedate = date_reference + datetime.timedelta(days=3, hours=22)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Thu
    task_name = u'프로그래밍언어구조론 시험 점수 확인'
    duedate = date_reference + datetime.timedelta(days=4, hours=16)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'영화의이해 보고서 작성'
    duedate = date_reference + datetime.timedelta(days=4, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'이산구조 과제 마감'
    duedate = date_reference + datetime.timedelta(days=4, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'저녁 약속'
    duedate = date_reference + datetime.timedelta(days=4, hours=19)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Fri
    task_name = u'만화책 구매'
    duedate = date_reference + datetime.timedelta(days=5, hours=16)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'홍대 코믹월드 예매권 구매하러 가기'
    duedate = date_reference + datetime.timedelta(days=5, hours=16)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'야구 경기 보러 가기'
    duedate = date_reference + datetime.timedelta(days=5, hours=18)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    ### Sat
    task_name = u'기차표 예매'
    duedate = date_reference + datetime.timedelta(days=6, hours=0)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'게임 예약 구매'
    duedate = date_reference + datetime.timedelta(days=6, hours=12)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'강남역 점심 식사'
    duedate = date_reference + datetime.timedelta(days=6, hours=12)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    task_name = u'소마 멘토링 발표 준비'
    duedate = date_reference + datetime.timedelta(days=6, hours=23)
    task_collection.append(createTaskInfo(user_id, task_name, duedate))
    
    return task_collection


tasks = genSampleTask()
```
